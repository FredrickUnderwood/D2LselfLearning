```python
CSV_PATH = './input/train.csv'
IMAGES_PATH = './input/train_images'
import torchvision
from PIL import Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

# 尝试部分

## 1. 突出绿色部分

```python
train_csv = pd.read_csv(CSV_PATH)
first_image_name = train_csv.iloc[0, 0]
first_image = Image.open(os.path.join(IMAGES_PATH, first_image_name))
first_image = np.array(first_image)
plt.imshow(first_image)
```

```python
ABNORMAL_IMAGE1_PATH = '159654644.jpg'
first_abnormal_image = Image.open(os.path.join(IMAGES_PATH, ABNORMAL_IMAGE1_PATH))
first_abnormal_image = np.array(first_abnormal_image)
plt.imshow(first_abnormal_image)
```

```python
lower_green = np.array([0, 50, 0])
upper_green = np.array([220, 255, 100])
mask = cv2.inRange(first_image, lower_green, upper_green)
result = cv2.bitwise_and(first_image, first_image, mask=mask)
plt.imshow(result)
```

```python
lower_green = np.array([0, 50, 0])
upper_green = np.array([180, 255, 60])
mask = cv2.inRange(first_abnormal_image, lower_green, upper_green)
result = cv2.bitwise_and(first_abnormal_image, first_abnormal_image, mask=mask)
plt.imshow(result)
```

### aug_test_1

```python
AUG_TEST_1_PATH = './aug_test_1'
lower_green = np.array([30, 50, 0])
upper_green = np.array([255, 255, 120])
for i in range(21):
    image_name = train_csv.iloc[i, 0]
    aug_test_1_image = Image.open(os.path.join(IMAGES_PATH, image_name))
    aug_test_1_image = np.array(aug_test_1_image)
    mask = cv2.inRange(aug_test_1_image, lower_green, upper_green)
    result = cv2.bitwise_and(aug_test_1_image, aug_test_1_image, mask=mask)
    result = Image.fromarray(result)
    result.save(os.path.join(AUG_TEST_1_PATH, image_name))
```
